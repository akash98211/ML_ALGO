```python
import pandas as pd
import numpy as np

from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, multilabel_confusion_matrix
```

```python
df = pd.read_csv('Iris.csv')
```

```python
# df
```

```python
df['Species'].value_counts()
```

```python
df['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2},inplace = True)
```

```python
df.head(3)
```

```python
x = df.drop(['Id','Species'],axis = 1)
y = df['Species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2 , random_state=10, stratify=y)
```

```python
ab_model = AdaBoostClassifier(random_state=1)
ab_model.fit(x_train, y_train)
```

# Evaluation

```python
# Evaluation on Test Data

y_pred = ab_model.predict(x_test)

test_accuracy = accuracy_score(y_test, y_pred)
print('Accuracy of model is :',test_accuracy)

clf_report = classification_report(y_test, y_pred)
print("Classification Report :\n",clf_report)
```

```python
# Evaluation on Train Data

y_pred_train = ab_model.predict(x_train)

train_accuracy = accuracy_score(y_train, y_pred_train)
print('Accuracy of model is :',train_accuracy)

clf_report = classification_report(y_train, y_pred_train)
print("Classification Report :\n",clf_report)

print(confusion_matrix(y_train, y_pred_train))

multilabel_confusion_matrix(y_train, y_pred_train)
```
