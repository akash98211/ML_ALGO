# Algorithm

```python
Best Fit Line >> Regression Line >> Model 

Calculate best m and c values:
    
Cost Function >> MSE 

Gradient Descent Algorithm:
    Used to minimize MSE(Cost Function)
    Partial Derivative
    
1. Initialize m and c values (random values)
2. Mnew = M - L(PD(MSE)/m)
3. Cnew = C - L(PD(MSE)/c)
Learning Rate: To avoid overshooting
4. Global Minima points:
    Lowest MSE
    MSE will increase if we change the value of m and c
    

Best Fit Line:
    It passes throght max number of datapoints
    
    
```

```python
Coef of determination
Goodness of best fit line

SSE  >> RSS >> sum((Ya-Yp)^2)
SST  >> Total Error >> sum((Ya-Ymean)^2)

R2 = 1 - SSE/SST
R2 = 1  >> SSE = 0
R2 = 0  >> SSE == SST
R2 = <1 or negative  >> SSE>SST

```

## 6. Adjusted R2 Value

```python
Increase only for good predictors

Adjusted R2 = 1 - (1-R2)*(N-1)/(N-p-1)
N >> Number of sample
p >> Number of predictors
```

# Assumptions of Linear Regression

```python
1. Linearity
2. No Multicollinearity
3. Normality of Residual 
4. Homoscedasticity of Residual or No Heteroscedacticity:
    Residual should follow Homoscedasctic behaviour

    Fitted values vs Residual
```

# Advantages

```python
1. Perform exceptionally well on linearly seperable data
2. Easy to implement 
3. Overfitting can be reduced by regularization(L1 and L2)
```

# Disadvantages

```python
1. Linearity
2. Independence
3. Sensitive to outliers
4. Sensitive to missing value
```

# Overfitting and Underfitting

```python

Overfitting:
    Train data accuracy  >> High  >> 96%  >> 1000
    Test Data accuracy   >> Low   >> 75%  >> 5000
    Low Bias and High Variance
    
    
Underftting: 
    Train data accuracy  >> Low >> 70
    Test Data accuracy   >> Low >> 70
    
    High Bias and Low Variance
    
Bias >> Accuracy on train data:
    Low Bias >> High Accuracy
    High Bias >> Low Accuracy
    
Variance:
    Difference between accuracies of different datasets
    
    High Variance >> 
    1. High Train Accuracy and Low Test accuracy  >> More difference
    2. Low Train Accuracy and High Testing Accuracy

    Low Variance:
        1. High Train Accuracy and High Test accuracy
        2. Low Train Accuracy and Low Testing Accuracy
```
