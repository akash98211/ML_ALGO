```python
# import libraries
import numpy as np
import pandas as pd

# decision tree libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

# spliting Data
from sklearn.model_selection import train_test_split

# Hyperparameter
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV

# Evaluation
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# warnings
import warnings
warnings.filterwarnings('ignore')
```

# 1. Data Gatering

```python
df=pd.read_csv("diabetes.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

# 2.selection feature

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
x.head()
```

```python
y.head()
```

# 3. model training

```python
x_train ,x_test, y_train ,y_test = train_test_split(x,y,test_size=0.2 , random_state=1, stratify=y)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# 4.model Building

```python
dt_model=DecisionTreeClassifier(criterion='gini')
dt_model.fit(x_train,y_train)
```

```python
# Accuracy on Testing Dataset

y_pred=dt_model.predict(x_test)

testing_accuracy=accuracy_score(y_test,y_pred)
print(" Testing Accuracy score is : ", testing_accuracy)

clf_report=classification_report(y_test,y_pred)
print("classification_report is : ",clf_report)


```

```python
# Accuracy on Training Dataset

y_pred_train=dt_model.predict(x_train)

training_accuracy=accuracy_score(y_train,y_pred_train)
print(" Testing Accuracy score is : ", testing_accuracy)

clf_report=classification_report(y_train,y_pred_train)
print("classification_report is : ",clf_report)


```

# 5.plot Decision Tree

```python
plt.figure(figsize=(200,100))
tree=plot_tree(decision_tree=dt_model,feature_names=df.columns,class_names=['NO','YES'],filled=True)
print("Decision Tree ploting ")
plt.savefig("Decision Tree ploting")
```

# 6. Hyperparameter Tunning

```python
dt_model=DecisionTreeClassifier(random_state=1)
dt_model.fit(x_train,y_train)
```

```python
Hyperparameter={'criterion':['gini','entropy'],
               'max_depth':np.arange(2,10),
                'min_impurity_split':np.arange(2,20),
                'min_samples_leaf':np.arange(1,15)
               }
```

```python
gscv_dt_model=GridSearchCV(dt_model,Hyperparameter, cv=5)
gscv_dt_model.fit(x_train,y_train)
```

```python
gscv_dt_model.best_estimator_
```

```python
dt_model=DecisionTreeClassifier(criterion='gini',random_state=1 ,max_depth=2, min_samples_split=2, min_samples_leaf=1)
dt_model.fit(x_train,y_train)
```

```python
# Accuracy on Testing Dataset

y_pred=dt_model.predict(x_test)

testing_accuracy=accuracy_score(y_test,y_pred)
print(" Testing Accuracy score is : ", testing_accuracy)

clf_report=classification_report(y_test,y_pred)
print("classification_report is : ",clf_report)


```

```python
# Accuracy on Training Dataset

y_pred_train=dt_model.predict(x_train)

training_accuracy=accuracy_score(y_train,y_pred_train)
print(" Testing Accuracy score is : ", testing_accuracy)

clf_report=classification_report(y_train,y_pred_train)
print("classification_report is : ",clf_report)


```
